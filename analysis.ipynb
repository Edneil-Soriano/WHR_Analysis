# Regression

We use different regression models to determine a mapping from individual features (Freedom, Economy, etc...) to the overall Happiness scores. We can then determine which feature is the most important on a person's happiness. 

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso, RidgeCV, LassoCV
from sklearn.preprocessing import StandardScaler

features = [
    'Economy (GDP per Capita)',
    'Family',
    'Health (Life Expectancy)',
    'Freedom',
    'Trust (Government Corruption)',
    'Generosity'
]

X = whr['2015.csv'][features]
y = whr['2015.csv']['Happiness Score']

# standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# linear reg
linreg = LinearRegression()
linreg.fit(X_scaled, y)
linreg_coefs = pd.Series(linreg.coef_, index=features)

# === First Subplot: Promoting Sparsity (fixed alpha) ===
ridge_sparse = Ridge(alpha=1.0)
ridge_sparse.fit(X_scaled, y)
ridge_sparse_coefs = pd.Series(ridge_sparse.coef_, index=features)

lasso_sparse = Lasso(alpha=0.5)
lasso_sparse.fit(X_scaled, y)
lasso_sparse_coefs = pd.Series(lasso_sparse.coef_, index=features)

df_sparse = pd.DataFrame({
    'Linear': linreg_coefs.abs(),
    'Ridge (α=1.0)': ridge_sparse_coefs.abs(),
    'Lasso (α=0.5)': lasso_sparse_coefs.abs()
}).sort_values(by='Linear', ascending=True)

# === Second Subplot: Following Cross-Validation ===
alphas = np.logspace(-4, 2, 100)

ridge_cv = RidgeCV(alphas=alphas, store_cv_values=True)
ridge_cv.fit(X_scaled, y)
ridge_cv_coefs = pd.Series(ridge_cv.coef_, index=features)

lasso_cv = LassoCV(alphas=alphas, cv=5, max_iter=10000)
lasso_cv.fit(X_scaled, y)
lasso_cv_coefs = pd.Series(lasso_cv.coef_, index=features)

print(f"Best alpha for Ridge (CV): {ridge_cv.alpha_:.5f}")
print(f"Best alpha for Lasso (CV): {lasso_cv.alpha_:.5f}")

df_cv = pd.DataFrame({
    'Linear': linreg_coefs.abs(),
    f'Ridge (α={ridge_cv.alpha_:.2g})': ridge_cv_coefs.abs(),
    f'Lasso (α={lasso_cv.alpha_:.2g})': lasso_cv_coefs.abs()
}).sort_values(by='Linear', ascending=True)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 6), sharey=True)

df_sparse.plot(kind='barh', ax=axes[0], color=['skyblue', 'salmon', 'lightgreen'])
axes[0].set_title('Importance of Happiness Metrics (promoting sparsity)')
axes[0].set_xlabel('Standardized Coefficients')

df_cv.plot(kind='barh', ax=axes[1], color=['skyblue', 'salmon', 'lightgreen'])
axes[1].set_title('Importance of Happiness Metrics (following cross-validation)')
axes[1].set_xlabel('Standardized Coefficients')

plt.tight_layout()
plt.show()
fig.savefig('figures/Regression.png')
```

# PCA 

Since we now determine what makes a person happy, we now want to know which people in the world are happy. In particular, we use PCA to partition countries based on 5 categories of happiness. Lastly, we also want to determine which features are correlated. Hence, if you are someone who wants to make people happy, a good feature to improve at first are those that are correlated since improving one will probably improve the other. 

```python
from sklearn.decomposition import PCA
```

```python
X = whr['2015.csv'][features].copy()
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)


new_labels = ['Economy', 'Family', 'Health', 'Freedom', 'Government', 'Generosity']

corr_df = pd.DataFrame(X_scaled, columns=new_labels)
plt.figure(figsize=(5, 4))
sns.heatmap(corr_df.corr(), annot=True, cmap='Reds')

plt.xticks(rotation=20, fontsize=8)
plt.yticks(rotation=0, fontsize=8)

plt.title('Correlation Matrix', fontsize=10)
plt.tight_layout()
plt.show()
plt.savefig('figures/corr_matrix.png')
```

```python
pca = PCA(n_components=6)
X_pca = pca.fit_transform(X_scaled)

explained_var_ratio = pca.explained_variance_ratio_
cumulative_var = np.cumsum(explained_var_ratio)
```

```python
plt.figure(figsize=(8, 4))
plt.step(range(1, 7), cumulative_var, where='mid',  color='red')
plt.xlabel('Principal Component')
plt.ylabel('Explained Variance (Normalized)')
plt.title('Scree Plot - 2 PCs can explain 70% of variance')
plt.xticks(range(1, 7))
plt.grid(True)
plt.tight_layout()
plt.show()
```

# K-means clustering in PC space

```python
from sklearn.cluster import KMeans

#KMeans on PCA features
kmeans = KMeans(n_clusters=2, random_state=42)
clusters = kmeans.fit_predict(X_pca)

df_pca = pd.DataFrame(X_pca, columns=[f'PC{i+1}' for i in range(6)])
df_pca['Country'] = whr['2015.csv']['Country']
df_pca['Cluster'] = clusters
df_pca['Happiness Score'] = whr['2015.csv']['Happiness Score']

cluster_mean_scores = df_pca.groupby('Cluster')['Happiness Score'].mean()
happy_cluster = cluster_mean_scores.idxmax()
df_pca['Mood'] = df_pca['Cluster'].apply(lambda x: 'Happy' if x == happy_cluster else 'Sad')
```

```python
plt.figure(figsize=(8, 6))


sns.scatterplot(x='PC1', y='PC2', hue='Mood', data=df_pca, palette={'Happy': 'skyblue', 'Sad': 'salmon'})
countries_to_label = ['Philippines', 'United States', 'Italy', 'China', 'Germany', 'Switzerland', 'Sudan', 
                      'South Africa', 'Ghana', 'Togo', 'Syria', 'Burundi', 'Myanmar']
highlight = df_pca[df_pca['Country'].isin(countries_to_label)]


plt.scatter(highlight['PC1'], highlight['PC2'], marker='.', s=100, c='black')

for i, row in highlight.iterrows():
    plt.text(row['PC1'] + 0.02, row['PC2'] + 0.02, row['Country'], fontsize=9)

plt.title('Clustering of Countries by Happiness Metrics (k=2)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(title='Cluster Mood')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python

kmeans = KMeans(n_clusters=5, random_state=42)
clusters = kmeans.fit_predict(X_pca)
df_pca = pd.DataFrame(X_pca, columns=[f'PC{i+1}' for i in range(6)])
df_pca['Country'] = whr['2015.csv']['Country']
df_pca['Cluster'] = clusters
df_pca['Happiness Score'] = whr['2015.csv']['Happiness Score']
cluster_mean_scores = df_pca.groupby('Cluster')['Happiness Score'].mean().sort_values(ascending=False)

friendly_labels = ['Very Happy', 'Happy', 'Neutral', 'Unhappy', 'Very Unhappy']
mood_labels = {cluster: friendly_labels[i] for i, cluster in enumerate(cluster_mean_scores.index)}
df_pca['Mood'] = df_pca['Cluster'].map(mood_labels)


plt.figure(figsize=(10, 8))
sns.scatterplot(x='PC1', y='PC2', hue='Mood', data=df_pca, palette='Set1', edgecolor='w', linewidth=0.5)


countries_to_label = [
    'Philippines', 'United States', 'France', 'China', 'Germany', 'Switzerland',
    'Sudan', 'South Africa', 'Ghana', 'Togo', 'Syria', 'Burundi', 'Myanmar'
]
highlight = df_pca[df_pca['Country'].isin(countries_to_label)]

plt.scatter(highlight['PC1'], highlight['PC2'], 
            facecolors='none', edgecolors='black', s=150, linewidths=1.5)

for _, row in highlight.iterrows():
    plt.text(row['PC1'] + 0.02, row['PC2'] + 0.02, row['Country'], fontsize=9)

plt.title('Clustering of Countries by Happiness Metrics (k=5)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(title='Cluster Mood')
plt.grid(True)
plt.tight_layout()
plt.show()
```
